```
import time
import random
from datetime import datetime
import pytz

print('start infinite loop')
timeZ_Jakarta = pytz.timezone('Asia/Jakarta')

counter = 0
while True:
  counter += 1
  # now = datetime.now()
  dt_Jakarta = datetime.now(timeZ_Jakarta)
  current_time = dt_Jakarta.strftime("%H:%M:%S")
  print("counter:", counter, "Current Time:", current_time)
  time.sleep(3)
```
